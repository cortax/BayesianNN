# Libraries

```python
import pystan
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import uniform, norm
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import torch
```

# Model definition

```python
model = """
data {
    int<lower=0> N;
    vector[N] x;
    vector[N] y;
}
parameters {
    real A;
    real b;
    real<lower=0> sigma;
}
model {
    y ~ normal(A * x + b, sigma);
}
"""
```

# Data generation

```python
np.random.seed(0)
A = 1.2
b = -0.3
sigma = 0.3
N = 200
x = torch.from_numpy(uniform.rvs(size=N))
x.detach()
x = x.float()
noise = torch.from_numpy(norm.rvs(loc=0, scale=sigma, size=N)).float()
y = A*x + b + noise
y.detach()
y = y.float()
plt.scatter(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatterplot')
```

```python
# Put our data in a dictionary
data = {'N': len(x), 'x': x, 'y': y}

# Compile the model
sm = pystan.StanModel(model_code=model)

# Train the model and generate samples
fit = sm.sampling(data=data, algorithm="NUTS", iter=500, warmup=250, thin=1, seed=42, n_jobs=-1)
print(fit)
```

```python
summary_dict = fit.summary()
df = pd.DataFrame(summary_dict['summary'], 
                  columns=summary_dict['summary_colnames'], 
                  index=summary_dict['summary_rownames'])

A_mean, b_mean = df['mean']['A'], df['mean']['b']

# Extracting traces
A = fit['A']
b = fit['b']
sigma = fit['sigma']
lp = fit['lp__']
```

```python
%matplotlib notebook
x_min, x_max = 0, 1
x_plot = np.linspace(x_min, x_max, 100)

# Plot a subset of sampled regression lines
for _ in range(2000):
  A_p, b_p = np.random.choice(A), np.random.choice(b)
  plt.plot(x_plot, A_p * x_plot + b_p, color='lightsteelblue',
           alpha=0.005 )

# Plot mean regression line  
plt.plot(x_plot, A_mean * x_plot + b_mean)
plt.scatter(x, y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitted Regression Line')
plt.xlim(x_min, x_max)
```
